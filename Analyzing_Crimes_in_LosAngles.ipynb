| `'Vict Descent'` | Victim's descent:<ul><li>`A` - Other Asian</li><li>`B` - Black</li><li>`C` - Chinese</li><li>`D` - Cambodian</li><li>`F` - Filipino</li><li>`G` - Guamanian</li><li>`H` - Hispanic/Latin/Mexican</li><li>`I` - American Indian/Alaskan Native</li><li>`J` - Japanese</li><li>`K` - Korean</li><li>`L` - Laotian</li><li>`O` - Other</li><li>`P` - Pacific Islander</li><li>`S` - Samoan</li><li>`U` - Hawaiian</li><li>`V` - Vietnamese</li><li>`W` - White</li><li>`X` - Unknown</li><li>`Z` - Asian Indian</li> |

```python
# Re-run this cell
#Â Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
crimes = pd.read_csv("crimes.csv", parse_dates=["Date Rptd", "DATE OCC"], dtype={"TIME OCC": str})
crimes.head()
```

```python
# The hour with highest frequency of crimes
crimes["HOUR OCC"] = crimes["TIME OCC"].str[:2].astype(int)
sns.countplot(x = "HOUR OCC", data = crimes)
plt.show()
```

```python
# Midday has largest volume of crime
peak_crime_hour = 12
```

```python
# The area with largest frequency of night crimes (10pm-3:59am)
night_crimes = crimes[crimes["HOUR OCC"].isin([22, 23, 0, 1, 2, 3])]
peak_night_crime_location = night_crimes['AREA NAME'].value_counts().index[0]
print("Most of crimes occured in {}.".format(peak_night_crime_location))
```

```python
# The number of crimes committed againist victims by age groups
age_labels = ["<18", "18-25", "26-34", "35-44", "45-54", "55-64", "65+"]
age_bins = [0, 17, 25, 34, 44, 54, 64, np.inf]
crimes["VictAge_group"] = pd.cut(crimes["Vict Age"], bins=age_bins, labels= age_labels)
victim_ages = crimes["VictAge_group"].value_counts()
print(f"{victim_ages}")
```

```python
# Visualize the number of crimes committed againist age group
sns.catplot(x = "VictAge_group", data = crimes, kind = "count")
plt.title("Number of Crimes Committed Againist Age Group")
plt.ylabel("Number of crimes")
plt.show()
```
